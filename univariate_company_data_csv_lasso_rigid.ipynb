```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv("/content/company_dataset.csv")
```

```python
data.head()
```

```python
data.corr()
```

```python
x = data.drop('Sales', axis=1).values
y = data['Sales'].values
```

```python
x_Tv=x[:,2]
```

```python
type(x_Tv),type(y)
```

```python
x_Tv.shape,y.shape
```

```python
x_Tv = x_Tv.reshape(-1,1)
y = y.reshape(-1, 1)
```

```python
x_Tv.shape, y.shape
```

```python
plt.scatter(x_Tv, y)
plt.xlabel('televisions')
plt.ylabel('Sales')
```

```python
 reg = linear_model.LinearRegression()
 reg.fit(x_Tv, y)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_pred = reg.predict(x_Tv)
print(f'MAE = {mean_absolute_error(y, y_pred)}')
print(f'MSE = {mean_squared_error(y, y_pred)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y, y_pred))}')
print(f'R_2 = {r2_score(y, y_pred)}')
```

```python
prediction_space = np.linspace(min(x_Tv), max(x_Tv)).reshape(-1,1)
```

```python
plt.scatter(x_Tv, y, color='blue')
plt.plot(x_Tv, reg.predict(x_Tv ), color='red', linewidth=3)
```

```python
plt.scatter(x_Tv, y, color='blue')
plt.plot(prediction_space, reg.predict(prediction_space), color='black', linewidth=3)
```

LASSO REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
reg = Lasso(alpha = 0.1)
reg.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_pred1 = reg.predict(x_train)
print(f'Performance of the model on training data :\n')
print(f'MAE = {mean_absolute_error(y_train, y_pred1)}')
print(f'MSE = {mean_squared_error(y_train, y_pred1)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y_train, y_pred1))}')
print(f'R_2 = {r2_score(y_train, y_pred1)}')
```

```python
y_pred2 = reg.predict(x_test)
print(f'Performance of the model on test data :\n')
print(f'MAE = {mean_absolute_error(y_test, y_pred2)}')
print(f'MSE = {mean_squared_error(y_test, y_pred2)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y_test, y_pred2))}')
print(f'R_2 = {r2_score(y_test, y_pred2)}')
```

```python
reg.score(x_test, y_test)
```

RIGID REGRESSION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
reg = Ridge(alpha = 0.1)
reg.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_pred1 = reg.predict(x_train)
print(f'Performance of the model on training data :\n')
print(f'MAE = {mean_absolute_error(y_train, y_pred1)}')
print(f'MSE = {mean_squared_error(y_train, y_pred1)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y_train, y_pred1))}')
print(f'R_2 = {r2_score(y_train, y_pred1)}')
```

```python
y_pred2 = reg.predict(x_test)
print(f'Performance of the model on test data :\n')
print(f'MAE = {mean_absolute_error(y_test, y_pred2)}')
print(f'MSE = {mean_squared_error(y_test, y_pred2)}')
print(f'RMSE = {np.sqrt(mean_squared_error(y_test, y_pred2))}')
print(f'R_2 = {r2_score(y_test, y_pred2)}')
```

```python
reg.score(x_test, y_test)
```
